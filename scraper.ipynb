Scrapes data from thecorrect.com.

Data scraping method:
* Our strategy for scraping medical questions is as follows:
    1. have a list with medium publications that publish often
    2. pick a year over which to scrape the data (we used 2020)
    3. for such selected year, we scraped all articles from the list of publications
    
* Based on what we see in loaded page we want, we extracted the following data:
    1. Inspect. In our case, the relevant information in the table is associated with the tag <p\>. 
    2. Filter. We use the find_all method to filter out all the questions on the URL.
    3. Create structured data. We used pandas package to build a table with a column of question #, questions from patients, and questions form providers.

```python
# Make a GET request to fetch the raw HTML content
# importing the libraries
from bs4 import BeautifulSoup
import requests
html_content = requests.get(url).text

# Parse the html content
soup = BeautifulSoup(html_content, "lxml")
print(soup.prettify()) # print the parsed data of html
```

```python
print(soup.title)
```

```python
for link in soup.find_all("a"):
    print("Inner Text: {}".format(link.text))
    print("Title: {}".format(link.get("title")))
    print("href: {}".format(link.get("href")))
```

```python
soup.find_all("td")
```

```python
for tag in soup.find_all(True):
    print(tag.name)
```

```python

# Create all_h1_tags as empty list
all_h1_tags = []

# Set all_h1_tags to all h1 tags of the soup
for element in soup.select('a'):
    all_h1_tags.append(element.text)

qdf = all_h1_tags    

#print(all_h1_tags)
```
